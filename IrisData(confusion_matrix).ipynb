```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn

from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
x = iris.data       #independent
x.shape
```

```python
y = iris.target     #dependent
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 1)
```

```python
mymodel = LogisticRegression()
mymodel.fit(x_train, y_train)
```

```python
y_pred = mymodel.predict(x_test)    #result = y_test
y_pred
```

Calculate using Confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm        #since ther are 3 classes target (1 is incorre3ct pred)
```

Calculate using accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)     (#97% accuracy)
```
