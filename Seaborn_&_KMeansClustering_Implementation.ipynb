```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.DataFrame({ 'x1':[12,20,28,18,29,33,24,45,43,51,52,55,53,54,61,64,69,52,6,62,45,60,59,8], 
                 'x2':[39,36,30,52,54,46,55,59,63,70,66,63,58,14,8,19,7,24,77,65,80,4,72,45]})     #converting a dict to data frame (x1 and x2 are 2 features)
```

```python
from sklearn.cluster import KMeans
```

```python
x = df
df.head()
```

```python
km = KMeans(n_clusters = 3)   #3clusters
#now fit your data into km
```

```python
km.fit(x)
```

```python
km.cluster_centers_       #gives 3 centroids
```

```python
a = km.labels_
a
```

```python
df['K Mean Predicted Label'] = km.labels_       #combining labels and data
```

```python
df.head()
```

#or You can combine like this

```python
a = pd.DataFrame(a)
a
merge = pd.concat([df, a], axis = 'columns')
merge
```

```python
labels = km.labels_
labels
```

##Visualization

```python
plt.scatter(df['x1'],df['x2'], 30, labels)      #labels gives the color of to same cluster and 30 is the size of data points

for i in km.cluster_centers_:
  plt.scatter(i[0], i[1], 100, marker = '*')    #100 is the size of centroid
plt.show()

plt.show()

#if you have many features (x1,c2,x3,x4,....,xn) then select any 2 features only
```

## Elbow method (to find the optimal number of centroids)

```python
wss = []
cluster_range = range(1,10)

for k in cluster_range:
  km = KMeans(n_clusters = k)     #n_cluster is num of clusters 
  km.fit(x)

  inertia = km.inertia_     #gives sum of squares of distances
  wss.append(inertia)

print(wss)
```

```python
plt.xlabel('k. here no. of clusters')
plt.ylabel('Sum of squared error')
plt.plot(cluster_range, wss)
plt.show()
```
