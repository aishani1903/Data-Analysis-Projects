Visualization using matplotlib

```python
x = ds['YearsExperience']
y = ds['Salary']
plt.scatter(x,y, marker = '+', color = 'r')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Spliting of Dataset into testing and training

```python
x = ds.iloc[:,:-1].values     #converted to numpy 2d array(features)
y = ds.iloc[:,1].values       #converted to numpy 1d array(labels)
print(x)
print(y)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=1/3,random_state=1)
```

creating simple linear model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()      #will solve your y=ax+b and then we have to fit our data to this
model.fit(xtrain,ytrain)      #shows error that it needs 2d array n not 1d
```

Prediction

```python
y_predict = model.predict(xtest)
```
