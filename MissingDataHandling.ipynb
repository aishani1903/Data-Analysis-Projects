```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("student_data.csv")
df
```

```python
df.fillna(0)
```

```python
new_df = df.fillna({
    'CS' : 0,
    'Math' : 0,
    'Name' : 'No Category'
})
new_df
```

```python
new_df = df.fillna(method='ffill')
new_df      #ffill - forward propagation: means every null value tames it previous value
```

```python
new_df = df.fillna(method = 'bfill')
new_df
```

```python
new_df = df.fillna(method = 'bfill', axis = 'columns') 
new_df
```

```python
new_df = df.interpolate()     #puts average values
new_df
```

```python
new_df = df.dropna()      #deletes all rows with null values
new_df
```

```python
new_df = df.dropna(how='all')     #drops only those rows with all null values
new_df
```

```python
new_df = df.dropna(thresh = 2)
new_df
```

```python
new_df = df.replace(np.nan, 999)
new_df
```

```python
new_df = df.replace(np.nan, {
    'CS':1,
    'Math':2,
    'Name':3 
})
new_df
```
